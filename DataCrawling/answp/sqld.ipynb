```python
## 문제 추출
from bs4 import BeautifulSoup
import re
import requests

# 웹 페이지 URL
url = "https://yunamom.tistory.com/265"

# 페이지에 GET 요청을 보냄
response = requests.get(url)

# 요청이 성공했는지 확인
if response.status_code == 200:
    # 페이지의 HTML 소스코드를 파싱
    soup = BeautifulSoup(response.content, "html.parser")
    
    # 페이지의 모든 텍스트를 가져옴
    all_text = soup.get_text()
    
    # '문제' 뒤의 모든 텍스트를 추출하는 정규 표현식
    pattern = re.compile(r'문제 \d+\.\s*(.*?)(?=\n)')
    
    # 정규 표현식을 이용해 모든 매칭되는 텍스트를 찾음
    problems = pattern.findall(all_text)

    # 문제 번호 저장
    qnumber = []
    for i in range(len(problems)):
        qnumber.append(i + 1)
    print(qnumber)

    # 전처리
    problems = '\n'.join(problems)
    problems = problems.replace("문제 ", "", 4)
    problems = problems.replace("정답확인🌼", "")
    
    # 결과 출력
    print(problems)
else:
    print("페이지를 가져올 수 없습니다.")
```

```python
## 보기 추출

from bs4 import BeautifulSoup
import re
import requests

# 웹 페이지 URL
url = "https://yunamom.tistory.com/265"

# 페이지에 GET 요청을 보냄
response = requests.get(url)

# 요청이 성공했는지 확인
if response.status_code == 200:
    # 페이지의 HTML 소스코드를 파싱
    soup = BeautifulSoup(response.content, "html.parser")
    
    # 페이지의 모든 텍스트를 가져옴
    all_text = soup.get_text()
    
    # '1)'부터 '문제'까지의 모든 텍스트를 추출하는 정규 표현식
    pattern = re.compile(r'\n1\)(.*?)문제', re.DOTALL)
    
    # 정규 표현식을 이용해 모든 매칭되는 텍스트를 찾음
    matches = pattern.findall(all_text)
    
    # 결과 출력
    for match in matches:
        print("1)", match.strip())
else:
    print("페이지를 가져올 수 없습니다.")
```

```python
## 정답 추출

from bs4 import BeautifulSoup
import re
import requests

# 웹 페이지 URL
url = "https://yunamom.tistory.com/265"

# 페이지에 GET 요청을 보냄
response = requests.get(url)

# 요청이 성공했는지 확인
if response.status_code == 200:
    # 페이지의 HTML 소스코드를 파싱
    soup = BeautifulSoup(response.content, "html.parser")
    
    # 페이지의 모든 텍스트를 가져옴
    all_text = soup.get_text()
    
    # '정답 :' 뒤의 숫자를 추출하는 정규 표현식
    pattern = re.compile(r'정답\s*:\s*(.+?)(?=\n)')
    
    # 정규 표현식을 이용해 모든 매칭되는 텍스트를 찾음
    answers = pattern.findall(all_text)
    
    # 작은따옴표 제거하고 구분자를 '|'로 설정하여 출력
    answers = '|'.join(answers)
    answers = answers.replace(',', '중복되지않을 의미없는 문자')
    answers = answers.replace('|', ',').replace('중복되지않을 의미없는 문자', '|')
    
    # 결과 출력
    print(answers)
else:
    print("페이지를 가져올 수 없습니다.")
```
